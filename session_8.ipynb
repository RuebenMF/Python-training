```python
# Now it's your go. Write a function, with an appropriate name
# that divides one number by the other. Bonus points if you can make the function 
# always divide the largest of the two numbers by the smallest.
# Get the function to print the answer

def fnDiv(x, y):
	if x > y:
		print(f"{x} divided by {y} is {x / y}")
	else:
		print(f"{y} divided by {x} is {y / x}")

fnDiv(2,10)
```

```python
# You can define what is produced by the function using return
def sum_prod(num1, num2):
	sum = num1 + num2
	prod = num1 * num2

	return sum, prod

sum_ans, prod_ans = sum_prod(20, 3)

print(f"sum is {sum_ans}")
print(f"prod is {prod_ans}")
```

```python
# write a function that returns x / y and x - y, 
# always dividing the biggest by the smallest number 
# and always taking the smallest fromt he biggest

def minus_div(x, y):
	if x >= y:
		big = x
		small = y
	else:
		big = y
		small = x
	return big / small, big - small


minus_div(10,2)
```

```python
# here we apply a function to each row in a dataframe. this is much faster than creating a for-loop to do the same thing.

import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildCharacteristics.csv')

def ethnic_check(row):
	if row[0] == "W":
		return "Yes"
	else:
		return "No"

df['Is white'] = df['Ethnicity'].apply(ethnic_check)

df
```

```python
# 'https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildIdentifiers.csv'
# write a function that calculates ages based on the data, and returns a new column in age range buckets
# 0-5, 6-11, 12-17, 18+

# import data
# write functions
# using conditionals
# chaqnge to date
# calculation on df columns
# making new column

from dateutil.relativedelta import relativedelta

df = pd.read_csv('https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildIdentifiers.csv')

df['PersonBirthDate_dt'] = pd.to_datetime(df['PersonBirthDate'], format="%Y-%m-%d", errors='coerce')

df['Age'] = df['PersonBirthDate_dt'].apply(lambda row: relativedelta(pd.to_datetime('today'), row).years)

def age_bracket(age):
	if age < 6:
		return "0-5"
	elif age < 12:
		return "6-11"
	elif age < 18:
		return "12-17"
	else:
		return "18+"

df["Age bracket"] = df['Age'].apply(age_bracket)

df
```

```python
# Will's example

import numpy as np
# 'https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildIdentifiers.csv'
# write a function that calculates ages based on the data, and returns a new column in age range buckets
# 0-5, 6-11, 12-17, 18+

# import data
# write functions
# using conditionals
# chaqnge to date
# calculation on df columns
# making new column

df = pd.read_csv('https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildIdentifiers.csv')

def age_buckets(row):
    bday = pd.to_datetime(row, format="%Y-%m-%d")
    today = pd.to_datetime('today')

    # age = (today - bday)/pd.Timedelta(365.25, 'D')

    if bday + pd.DateOffset(years=6) > today:
        return '0-5 years'
    elif bday + pd.DateOffset(years=12) > today:
        return '6-11 years'
    elif bday + pd.DateOffset(years=18) > today:
        return '12-17 years'
    else:
        return '18+'
        
df['Age Bucket'] = df['PersonBirthDate'].apply(age_buckets)

df[['PersonBirthDate', 'Age Bucket']]
```
