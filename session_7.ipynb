```python
evens = list(range(2,13,2))
odds = list(range(1,13,2))

multiply_list = []

for even in evens:
	for odd in odds:
		result = even * odd
		statement = f'{even} times {odd} is {result}'
		multiply_list.append(statement)

print(multiply_list)
```

```python
# fizzbang
# take all the numbers from 1-100,
# print them, BUT if the number is divisible by 3, print fizz,
# if it's divisible by 5 print bang
# BUT if it's divisible by 3 & 5 print fizzbang
```

```python
# My Solution

fixbang_numbers = list(range(1,101))

for i in fixbang_numbers:
	if i % 15 == 0:
		print(f'fizzbang ({i})')
	elif i % 3 == 0:
		print(f'fizz ({i})')
	elif i % 5 == 0:
		print(f'bang ({i})')
	else:
		print(i)
```

```python
# William's Solution (1)

nums = list(range(1,101))

for num in nums:
	if (num % 3 == 0) & (num % 5 == 0):
		print('fizzbang') 
	elif num % 3 == 0:
		print('fizz')
	elif num % 5 == 0: 
		print('bang')
	else:
		print(num)
```

```python
# William's Solution (2)

nums = list(range(1,101))

for num in nums:
	if (num % 3 == 0):
		if (num % 5 == 0):
			print('fizzbang')
		else:
			print('fizz') 
	elif num % 5 == 0: 
		print('bang')
	else:
		print(num)
```
