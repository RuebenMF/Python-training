```python
#pandas
import pandas as pd #import the package 'pandas' and give it the alias 'pd'

filepath = r"https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/1980%202023%20average%20house%20prices.csv" #the r tells Python to treat the string as a raw string, and ignore characters like %

house_prices = pd.read_csv(filepath) #read the CSV file from the URL into a pandas DataFrame. If reading an excel file use pd.read_excel()

#house_prices.info() #get a summary of the DataFrame, including the number of non-null entries and data types for each column
house_prices["Period"].info() #get info on a specific column

house_prices.head() #print the first 5 rows of the DataFrame. 5 is default.
house_prices.tail(10) #print the last 10 rows of the DataFrame

type(house_prices) #check the type of the object, should be a pandas DataFrame

house_prices["Period"] #access a specific column in the DataFrame

mean_price = house_prices["Average price All property types"].mean() #calculate the mean of a specific column

condition = house_prices["Average price All property types"] < mean_price #create a condition to filter the DataFrame
less_than_mean = house_prices[condition] #apply the condition to the DataFrame to get rows where the condition is True

print(less_than_mean)
```

```python
import pandas as pd

filepath = r"https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildIdentifiers.csv"

ci = pd.read_csv(filepath)

type(ci)
#ci.info()
ci.tail(10)
ci.head()

ci["PersonBirthDate"].info()

#create a new column by copying an existing one
ci["PersonBirthDate_dt"] = ci["PersonBirthDate"]

# format argument for datetimes strftimes
ci["PersonBirthDate_dt"] = pd.to_datetime(ci["PersonBirthDate"], format="%Y-%m-%d") #convert the column to datetime format
#ci["PersonDeathDate_dt"] = pd.to_datetime(ci["PersonDeathDate"], format="%Y-%m-%d") #convert the column to datetime format
ci["ExpectedPersonBirthDate_dt"] = pd.to_datetime(ci["ExpectedPersonBirthDate"], format="%Y-%m-%d") #convert the column to datetime format


ci.head()
```

```python
# sum, min, max, median, mean
ciIDsum = ci["Unnamed: 0"].sum()
ciIDmin = ci["Unnamed: 0"].min()
ciIDmax = ci["Unnamed: 0"].max()
ciIDmedian = ci["Unnamed: 0"].median()
ciIDmean = ci["Unnamed: 0"].mean()

collection_end = pd.to_datetime("2026-03-31", format="%Y-%m-%d") #create a datetime object for comparison

ci["Age_delta"] = collection_end - ci["PersonBirthDate_dt"] #calculate the age at collection end date

ci["Age_delta"] = ci["Age_delta"] / pd.Timedelta(365.25, "d") #convert the age from timedelta to years

ci["Age_delta"] = ci["Age_delta"].astype(int)

ci_age_mean = ci["Age_delta"].mean()
ci_age_median = ci["Age_delta"].median()
ci_age_min = ci["Age_delta"].min()
ci_agi_max = ci["Age_delta"].max()
ci_age_sum = ci["Age_delta"].sum()

print(f"The mean age is {ci_age_mean}, the median age is {ci_age_median}, the minimum age is {ci_age_min}, the maximum age is {ci_agi_max}, and the sum of ages is {ci_age_sum}.")

```
