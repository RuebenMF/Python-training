```python
from sqlalchemy import (create_engine, 
						inspect, 
						text, 
						select, 
						MetaData, 
						Table, 
						and_,
						or_,
						desc,
						asc,
						func,
						)
 
gravity_engine = create_engine("sqlite+pysqlite://///workspaces/Python-training/Intermediate/session_1/data/gravity.db")

gravity_connection = gravity_engine.connect()
gravity_inspection = inspect(gravity_connection)

gravity_inspection.get_table_names()
```

```python
metadata_903 = MetaData()

episodes = Table('episodes', metadata_903, autoload_with=engine_903)
print(episodes.name)
print(episodes.c.keys())
print(repr(episodes))
```

```python
gravity_metadata = MetaData()

book = Table('book', gravity_metadata, autoload_with=gravity_engine)
print(book.name)
print(book.c.keys())
print(repr(book))
```

```python
with engine_903.connect() as con:
	stmt = "select * from episodes"
	result_proxy = con.execute(text(stmt))
	results = result_proxy.fetchall()
	
first_row = results[0]
print(first_row)
```

```python
with engine_903.connect() as con:
	stmt = select(episodes)
	# print(stmt)
	result_proxy = con.execute(stmt)
	results = result_proxy.fetchmany(size=5)

print(results)
```

```python
with gravity_engine.connect() as con:
	stmt = select(book)
	# print(stmt)
	result_proxy = con.execute(stmt)
	results = result_proxy.fetchmany(size=5)

print(results)
```

```python
header = Table('header', metadata_903, autoload_with=engine_903)

with engine_903.connect() as con:
	stmt = select(header).where(or_
								(header.columns.ETHNIC == "REFU",
								 header.columns.ETHNIC == "NOBT"))
	# print(stmt)
	result_proxy = con.execute(stmt)
	results = result_proxy.fetchmany(size=10)

print(results)
```

```python
# select all of the book table where the book ID is either 1 or 2

with gravity_engine.connect() as con:
	stmt = select(book).where(or_
								(book.columns.book_id == 1,
								 book.columns.book_id == 2))
	# print(stmt)
	result_proxy = con.execute(stmt)
	results = result_proxy.fetchmany(size=10)

print(results)
```
