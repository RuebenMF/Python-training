```python
from sqlalchemy import (create_engine, 
                        inspect, 
                        text, 
                        select, 
                        MetaData, 
                        Table, 
                        and_,
                        or_,
                        desc,
                        asc,
                        func,
                        )
 
gravity_engine = create_engine("sqlite+pysqlite://///workspaces/Python-training/Intermediate/session_1/data/gravity.db")

gravity_connection = gravity_engine.connect()
gravity_inspection = inspect(gravity_connection)

gravity_inspection.get_table_names()
```

```python
metadata_903 = MetaData()

episodes = Table('episodes', metadata_903, autoload_with=engine_903)
print(episodes.name)
print(episodes.c.keys())
print(repr(episodes))
```

```python
metadata_gravity = MetaData()

book = Table('book', metadata_gravity, autoload_with=gravity_engine)
print(book.name)
print(book.c.keys())
print(repr(book))
```

```python
with engine_903.connect() as con:
    stmt = "select * from episodes"
    result_proxy = con.execute(text(stmt))
    results = result_proxy.fetchall()
    
first_row = results[0]
print(first_row)
```
