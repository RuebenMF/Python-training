```python
# In this cell I've initialised a variable 'slice_this_phrase'. 
# On the next three lines, print the words dog, cat, and racoon seprately using a slice from the beginning, a slice giving a range in the middle, and a slice from the end.
# This means one slice will involve [:n], one will involve [n1:n2], and one will involve [-n:].

slice_this_phrase = 'dog cat racoon'

dog = slice_this_phrase[0:3]
cat = slice_this_phrase[4:7]
racoon = slice_this_phrase[-6:]

print(dog, cat, racoon)
```

```python
#Lets have a look at lists now. 
# We can initialise a list by using square brackets and assigning them to a variable. 
# We seperate elements of a list with commas. We can put pretty much whatever we want in a list: ints, strings, dictionaries, other lists, a whole range of things. 
# A full list of list methods can be found here: https://docs.python.org/3/tutorial/datastructures.html


#list_1 = [] # Even though there's nothing in here, it's a list, we can check by printing type(list_1)
#print(type(list_1))

colours_list = ['red', 'blue', 'green']
print(colours_list)

colours_list.append('purple') # The append method adds whatever is being appended to a list as one item in a list
print(colours_list)

extra_colours = ['turquoise', 'teal']
#colours_list.append(extra_colours)
print(colours_list) # You can see here it's appended the whole list extra_colours as one item in the list

# Instead, we can extend to have each element of a list as a new element in another list with the .extend() method
colours_list.extend(extra_colours)
print(colours_list)

# We can also access elements in a list the same way we can strings
print(colours_list[3:-2])

# We can remove list elements using.remove
colours_list.remove('green')
print(colours_list)
```

```python
# In this cell, make a list of a few foods and print it.
# Then extend the list with another food and print it.
# Next, make another list of foods, and use the extend method to add those extra foods as individual elements of the list, and print it.
# Finally, remove and element from the list and print the final list.
# Remember to always use a suitable name for your lists!

burger = ['bun','patty','cheese','onions','lettuce','tomato']
print(burger)

burger.append('bun')
print(burger)

additionaltoppings = ['relish','pickles','sauce']
burger.extend(additionaltoppings)
print(burger)

burger.remove('pickles')
print(burger)
```

```python
numbers_list = [2,3,4,5,1,6]
numbers_list.sort() # we can sort lists too, here using a method

print(numbers_list)

numbers_list.sort(reverse=True)
print(numbers_list)

sorted_list = sorted([3,2,3,4,5]) # Here we sort a list using a function and returning it to a new variable

colours_list = ['red', 'blue', 'Green', 'amber']
sorted_colours = sorted(colours_list, key=str.lower) #This is useful when sorting strings in a case-insensitive manner because uppercase letters come before lowercase letters in ASCII order
print(sorted_colours)
```

```python
# Now lets look at dictionaries. dictionaries may initially seem a little confusing, and not overly helpful, but they are an incredibly powerful tool.
# Dictionaries allow us to store things in key:value pairs, where, generally, the key is a string or a number, keys must be unique within a dictionary and must be immutable objects
# (objects that can't be changed, like lists), and the value is... pretty much whatever we want it to be.
# We build a list using curly brackets {}, seperating key:value pairs with colons, and entries with commas.

favourite_colours = {'Will':'teal',
					'Andy':['Purple', 'Green', 'Azure'],
					'Annie':{'Mondays':'Red',
							'Sundays':'Orange',
							'Other days':'Black'}}

print(favourite_colours) # Here you can see I've stored strings, lists, and even entire dictionaries inside strings
print(list(favourite_colours)) # list(d) will print the keys, in order from a dictionary
print(favourite_colours.items()) # Prints all key:value pairs
print(favourite_colours.keys()) # Prints all keys
print(favourite_colours.values()) # Prints all values
```

```python
# In this cell, make a dicitonary of some restaurants and the foods they sell, print the items, keys, and values, have at least one have the value as a list.

restaurants = {'McDonalds':['Big Mac','Mayo Chicken','Fries'],
				'Pizza Hut':'Pizza',
				'KFC':['Chicken Wings','Popcorn Chicken','Chicken drumsticks']
				}

print(restaurants)
print(restaurants.items())
print(restaurants.keys())
print(restaurants.values())
```

```python
print(favourite_colours['Will']) # We can access the values associated with a key by passing it between square brackets after the dictionary's name

favourite_colours['Naiomi'] = 'Yellow' # We can also add/update key value pairs in a similar way
print(favourite_colours)

print('Andy' in favourite_colours) # We can check if a key is in a dict or not like so
```
